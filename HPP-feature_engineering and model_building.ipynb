### feature engineering

```python
# handling missing values

feature_missing=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=="O"]
```

```python
# to check the percent of missing values

for feature in feature_missing:
    print('{}:{} % missing values'.format(feature,round(df[feature].isnull().mean(),4)))
```

```python
df[feature_missing].head()
```

```python
# replacing missing values with new label

def replace_cat_feature(df,feature_missing):
    data=df.copy()
    data[feature_missing]=data[feature_missing].fillna('missing')
    return data
```

```python
df=replace_cat_feature(df,feature_missing)
df[feature_missing].isnull().sum()
```

```python
# numerical values containing missing values

num_feature_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!="O"]
```

```python
for feature in num_feature_nan:
    print('{}:{} % missing values'.format(feature,round(df[feature].isnull().mean(),4)))
```

```python
# replacing missing values in numerical columns

for feature in num_feature_nan:
    median_value=df[feature].median()
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    
    df[feature].fillna(median_value,inplace=True)

df[num_feature_nan].isnull().sum()
```

```python
# the missing values for both the numerical and categorical was treated
```

```python
df.columns
```

```python
# dealing with temporal variable

for feature in ['YearBuilt', 'YearRemodAdd','GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
```

```python
df[['YearBuilt', 'YearRemodAdd','GarageYrBlt']].head()
```

```python
# treating skewness using log transformation

num_features=['LotFrontage','LotArea','GrLivArea','SalePrice']
```

```python
for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
# Handling rare categorical values
categorical_feature=[feature for feature in df.columns if df[feature].dtypes=='O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'rare_var')
```
