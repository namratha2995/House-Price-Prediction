### EDA

- checking for missing values
- checking for numerical variables
- checking for the distribution of the numerical variables
- categorical variables
- types of categorical variables

```python
# checking the presence of missing values

missing_value_feature=[feature for feature in df.columns if df[feature].isnull().sum()>1]
```

```python
df[missing_value_feature].head()
```

```python
# printing the list of missing value columns

for feature in missing_value_feature:
    print(feature,np.round(df[feature].isnull().mean(),4),"% missing values")
```

```python
# finding the relationship with missing value columns and dependent variable

for feature in missing_value_feature:
    data=df.copy()
    df[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
print('Id of houses {} '.format(len(df['Id'])))
```

```python
# exploring numerical feature

numerical_columns=[feature for feature in df.columns if df[feature].dtypes!="O"]
```

```python
# Inspect for the numerical features in the dataset

print('Number of numerical variables', len(numerical_columns))
```

```python
df[numerical_columns].head()
```

```python
# after clear examination we can find the presence of year column in numerical feature which has to converted into date index
# lising the year columns

year_feature=[feature for feature in numerical_columns if 'Yr' in feature or 'Year' in feature]
```

```python
# exploring the year variable

for feature in year_feature:
    print(feature, df[feature].nunique())
```

```python
# relation of year columns with dependent variable

df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Sales Price')
plt.title('Year Sold v/s Sales Price')
```

```python
# sales price is decreasing as year goes which is very unusual
```

```python
# to find rest of the year columns with sales price

for feature in year_feature:
    if feature !='YrSold':
        data=df.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Salesprice')
        plt.show()
```

```python
# Exploring the discrete numerical features

discrete_features=[feature for feature in numerical_columns if len(df[feature].unique())<25 
                   and feature not in year_feature+['Id']]
```

```python
print('Count of discrete variable: {}'.format(len(discrete_features)))
```

```python
df[discrete_features].head()
```

```python
# relationship of discrete feature with sales price

for feature in discrete_features:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```

```python
# exploring continous variables

continous_features=[feature for feature in numerical_columns if feature not in discrete_features+year_feature+['Id']]
```

```python
print('Count of continious variable',len(continous_features))
```

```python
df[continous_features].head()
```

```python
# finding the relation between continous feature and Sales price

for feature in continous_features:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# Almost all the features are right skewed
```

```python
# logarithmic transformation

for feature in continous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sales Price')
        plt.title(feature)
        plt.show()
```

```python
# all the above are positive correlation
```

```python
# checking for outliers

for feature in continous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
# the above features have outliers and we have to treat it
```

```python
# exploring categorical variable

categorical_features=[feature for feature in df.columns if df[feature].dtypes=="O"]
```

```python
print('Count of categorical variable',len(categorical_features))
```

```python
df[categorical_features].head()
```

```python
# finding the relation between categorical features and sales price

for feature in categorical_features:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sales Price')
    plt.title(feature)
    plt.show()
```
